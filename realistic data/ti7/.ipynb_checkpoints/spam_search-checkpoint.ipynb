# Read data and make team, assign players
1. finish build team
2. finish build player_consist_of

```python
import csv
team_name = []
with open('dota2ti7team.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        rows = row[0].split('\t')
        team_name.append(rows[0])
team_name = team_name[1:]
```

```python
import pandas as pd
```

```python
# eliminate player in two teams
dict = {}
players = []
player_in_team = []
new_team = []
position = []
cnt = 0;
for idx,name in enumerate(team_name):
    if (os.path.exists('useteam/' + name + '.csv')):
        df = pd.read_csv('useteam/' + name + '.csv')
        player = df.Player[1:6].tolist()
        new_team.append(name)
        dict[name] = []
        position_cnt = 0;
        for p in player:
            cnt = 6
            while (p in players):
                p = df.Player[cnt]
                cnt += 1
            dict[name].append(p)
            players.append(p)
            # add idx of the team
            player_in_team.append(len(new_team))
            # add position
            position.append(position_cnt % 5 + 1)
            position_cnt += 1
    
```

```python
len(new_team)
```

```python
# see if there is duplciates
dupl = []
for (idx1,p1) in enumerate(players):
    for (idx2, p2) in enumerate(players):
        if (p1 == p2 and idx1 != idx2):
            dupl.append(p1)
```

```python
pd_consist_of = pd.DataFrame({'player_id': range(1, 81), 
                            'player_position': position,
                            'player_team_id' : player_in_team,
                            'player_name': players})
```

# make participate.csv

```python
Ti_league = {}
for i in range(2, 7):
    Ti_league[str(i)] = []
    team_name_append = []
    with open('../ti' + str(i) +'/201' + str(i) + '.csv', 'r') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
        for row in spamreader:
            rows = row[0].split('\t')
            team_name_append.append(rows[0])
        team_name_append = team_name_append[1:]
        for name in team_name_append:
            if name in new_team:
                Ti_league[str(i)].append(name)
Ti_league[str(7)] = new_team
```

```python
a21 = (2011,8,1,0,0,0,0,0,0)
a22 = (2012,7,6,23,59,59,0,0,0)
a31 = (2012,8,1,0,0,0,0,0,0)
a32 = (2013,7,6,23,59,59,0,0,0)
a41 = (2013,8,1,0,0,0,0,0,0)
a42 = (2014,7,6,23,59,59,0,0,0)
a51 = (2014,8,1,0,0,0,0,0,0)
a52 = (2015,7,6,23,59,59,0,0,0)
a61 = (2015,8,1,0,0,0,0,0,0)
a62 = (2016,7,6,23,59,59,0,0,0)
a71 = (2016,8,1,0,0,0,0,0,0)
a72 = (2017,7,6,23,59,59,0,0,0)
A1 = [a21, a31, a41, a51, a61, a71]
A2 = [a22, a32, a42, a52, a62, a72]
```

```python
date = []
tm = []
Match = []
league = []
```

```python
match_count = [0, 0, 18, 81, 213, 369, 551, 791]
```

```python
for i in range(2, 8):
    start=time.mktime(A1[i - 2])
    end=time.mktime(A2[i - 2])
    for j in range(match_count[i - 1] + 1, match_count[i] + 1):
        t=random.randint(start, end)
        date_touple=time.localtime(t)
        date.append(time.strftime("%Y-%m-%d",date_touple))
        tm.append(time.strftime("%H:%M:%S", date_touple))
        Match.append(j)
        league.append(i)
        
        
```

```python
len(Match)
```

```python
df_match = pd.DataFrame({'Match' : Match,
                         'league' : league,
                         'date': date,
                         'tm' : tm
                        })
```

# make team_compose.csv

```python
random.seed(100)
```

```python
tc_compose = []
tc_radiant = []
tc_fb = []
tc_win = []
tc_tower = []
for i in Match:
    fb_rd = random.randint(0, 1)
    win_rd = random.randint(0, 1)
    tc_compose.append(i)
    tc_radiant.append(0)
    tc_fb.append(fb_rd)
    tc_win.append(win_rd)
    tc_tower.append(random.randint(0, 11))
    
    tc_compose.append(i)
    tc_radiant.append(1)
    tc_fb.append(fb_rd ^ 1)
    tc_win.append(win_rd ^ 1)
    tc_tower.append(random.randint(0, 11))
```

```python
df_tc = pd.DataFrame({'tc_compose' : tc_compose,
                         'tc_radiant' : tc_radiant,
                         'tc_fb': tc_fb,
                         'tc_win' : tc_win,
                          'tc_tower' : tc_tower
                        })
```

```python
df_tc = df_tc[['tc_compose', 'tc_radiant', 'tc_fb', 'tc_win', 'tc_tower']]
```

```python
df_tc
```

```python
df_tc.to_csv('output/team_compose.csv', encoding='utf-8', index = False, header=False)
```

```python
random.seed(100)
```

```python
pi_compose = []
pi_radiant = []
pi_team = []
for i in Match:
    league_number = 0;
    if (i >= 1 and i <=18):
        league_number = 2
    elif i <= 81:
        league_number = 3
    elif i <= 213:
        league_number = 4
    elif i <= 369:
        league_number = 5
    elif i <= 551:
        league_number = 6
    elif i <= 791:
        league_number = 7
    length = len(Ti_league[str(league_number)])
    rd1 = 0
    rd2 = 0
    while rd1 == rd2:
        rd1 = random.randint(0, length - 1)
        rd2 = random.randint(0, length - 1)
    pi_compose.append(i)
    pi_radiant.append(0)
    pi_team.append(new_team.index(Ti_league[str(league_number)][rd1]))
    
    pi_compose.append(i)
    pi_radiant.append(1)
    pi_team.append(new_team.index(Ti_league[str(league_number)][rd2]))
```

```python
df_Play_In = pd.DataFrame({'pi_compose' : pi_compose,
                           'pi_radiant' : pi_radiant,
                           'pi_team' : pi_team
                            })
```

```python
df_Play_In
```

```python
df_Play_In.to_csv('output/play_in.csv', encoding='utf-8', index = False, header=False)
```
