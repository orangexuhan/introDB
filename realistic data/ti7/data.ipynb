# Read data and make team, assign players

```python
import csv
team_name = []
with open('dota2ti7team.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        rows = row[0].split('\t')
        team_name.append(rows[0])
team_name = team_name[1:]
```

```python
import pandas as pd
```

```python
# eliminate player in two teams
dict = {}
players = []
player_in_team = []
new_team = []
position = []
cnt = 0;
for idx,name in enumerate(team_name):
    if (os.path.exists('useteam/' + name + '.csv')):
        df = pd.read_csv('useteam/' + name + '.csv')
        player = df.Player[1:6].tolist()
        new_team.append(name)
        dict[name] = []
        position_cnt = 0;
        for p in player:
            cnt = 6
            while (p in players):
                p = df.Player[cnt]
                cnt += 1
            dict[name].append(p)
            players.append(p)
            # add idx of the team
            player_in_team.append(len(new_team))
            # add position
            position.append(position_cnt % 5 + 1)
            position_cnt += 1
    
```

```python
len(new_team)
```

```python
# see if there is duplciates
dupl = []
for (idx1,p1) in enumerate(players):
    for (idx2, p2) in enumerate(players):
        if (p1 == p2 and idx1 != idx2):
            dupl.append(p1)
```

```python
pd_consist_of = pd.DataFrame({'player_id': range(1, 81), 
                            'player_position': position,
                            'player_team_id' : player_in_team,
                            'player_name': players})
```

```python
pi_compose = []
pi_radiant = []
pi_team = []
for i in Match:
    league_number = 0;
    if (i >= 1 and i <=18):
        league_number = 2
    elif i <= 81:
        league_number = 3
    elif i <= 213:
        league_number = 4
    elif i <= 369:
        league_number = 5
    elif i <= 551:
        league_number = 6
    elif i <= 791:
        league_number = 7
    length = len(Ti_league[str(league_number)])
    rd1 = 0
    rd2 = 0
    while rd1 == rd2:
        rd1 = random.randint(0, length - 1)
        rd2 = random.randint(0, length - 1)
    pi_compose.append(i)
    pi_radiant.append(0)
    pi_team.append(new_team.index(Ti_league[str(league_number)][rd1]) + 1)
    
    pi_compose.append(i)
    pi_radiant.append(1)
    pi_team.append(new_team.index(Ti_league[str(league_number)][rd2]) + 1)
```
